2. Escriba una clase iterable que tenga la siguiente funcionalidad

   2.1 Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.

   2.2. Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
   
   2.3 Entregue un error si los datos se han agotado


```python
import pandas as pd

iris = pd.read_csv('iris.csv')
iris.drop(['Unnamed: 0'], axis=1,inplace=True)
iris.head()
```

```python
len(iris)
```

```python
class ClaseIterable:
    def __init__(self, dataframe):
        self.limit = len(dataframe)
        self.val = 0
        self.dataframe = dataframe
    
    # Hace esta clase iterable
    def __iter__(self):
        return self
    
    def __getitem__(self, n):  # accede a cada uno de los elementos para índices empezando desde 0.
        if n >= (self.limit-1):
            raise ValueError("¡Los datos se han agotado!")
        else:
            return self.dataframe[n:n+2]
    
    # Hace esta clase un iterador
    def __next__(self):
        if self.val > self.limit:
            raise ValueError("¡Saldo muy pequeño!")
        else:
            return_val = self.val
            self.val += 1
            return self.dataframe[return_val:self.val+1]
```

```python
a=ClaseIterable(iris)
print(a)
```

```python
for i in ClaseIterable(iris):
    print(i)
```
